Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Load dataset into dataframe

```python
credit_card_data = pd.read_csv("/content/creditcard.csv")
```

```python
# distribution of legit transaction & fradulent transaction
credit_card_data["Class"].value_counts()
```

This dataset is highly unbalanced

0 --> Normal Transaction

1 --> Fradulent Transaction

```python
# seperation of dataset for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

```python
# statistical measure of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transaction
credit_card_data.groupby('Class').mean()
```

Modify the dataset to make the dataset more similar

```python
legit_sample = legit.sample(n=3)
```

Concatenate two dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

Splitting the data into features and target

```python
X = new_dataset.drop(columns="Class", axis=1)
```

```python
Y = new_dataset['Class']
```

Split the data into training and testing model

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(Y.shape, Y_train.shape, Y_test.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuraccy on train data: ", training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy of testing data: ", testing_data_accuracy)
```
