Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

Data collection and preprocessing

```python
raw_mail_data = pd.read_csv("/content/mail_data.csv")
```

```python
raw_mail_data.head()
```

```python
raw_mail_data.tail()
```

```python
# replace the null values with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)), '')
```

```python
# printing the first five rows of the dataframe
mail_data.head()
```

```python
mail_data.shape
```

```python
# label encoding ham -> 1 and spam -> 0

mail_data.loc[mail_data['Category'] == "spam", 'Category'] = 0
mail_data.loc[mail_data['Category'] == "ham", 'Category'] = 1
```

```python
#spam will he represented my 0 and ham by 1

mail_data.head()
```

```python
# seperating the data by texts and label

X = mail_data["Message"]
Y = mail_data["Category"]
```

```python
X.head()
```

```python
Y.head()
```

Splitting the data  into training data and test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Feature Extraction

```python
# transform the text data to feature vectors that can be used as input to the Logistic regression

feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers

Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
# predict on the training data

predict_training_model = model.predict(X_train_features)
accuracy_training_model = accuracy_score(Y_train, predict_training_model)
```

```python
print("Accuracy of the train model: ", accuracy_training_model)
```

```python
# predict on the testing model

predict_testing_model = model.predict(X_test_features)
accuracy_testing_model = accuracy_score(Y_test, predict_testing_model)
```

```python
print("Accuracy of the test model: ", accuracy_testing_model)
```

Building the predict system using the current data

```python
input_mail = ["Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"]

input_mail_features = feature_extraction.transform(input_mail)

prediction = model.predict(input_mail_features)

print(prediction)

if(prediction[0] == 0):
  print("Spam Email")
else:
  print("Ham email")
```
